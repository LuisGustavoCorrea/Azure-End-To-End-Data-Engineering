## Doing Transformation for all tables ( Changing column names )

```
#Create list variable with the name tables
table_name = []
path_silver= 'mnt/silver/SalesLT'
#Loop for each table
for i in dbutils.fs.ls('mnt/silver/SalesLT'):
    table_name.append(i.name.split('/')[0])
#Show table_name list
table_name
```

```
from pyspark.sql import SparkSession

# Inicializando SparkSession
spark = SparkSession.builder.appName("TransformarNomesColunas").getOrCreate()

# Exemplo de DataFrame
input_path= '/mnt/silver/SalesLT/Address'
df = spark.read.format('delta').load(input_path)

# Função para transformar os nomes das colunas
def transform_column_name(name):
    import re
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1)

# Obter os novos nomes das colunas
new_column_names = [transform_column_name(col) for col in df.columns]

# Renomear as colunas do DataFrame
df = df.toDF(*new_column_names)

# Mostrar o DataFrame com os novos nomes de colunas
display(df)
```
